```python
total_cost = int(input("Cost of the dream home: "))

portion_down_payment = 0.25
down_payment = total_cost * portion_down_payment
print("Down payment: ", down_payment)
```

```python
current_savings = 0

r = 0.04

anual_salary = int(input("Annual salary: "))
monthly_salary = anual_salary / 12
portion_saved = float(input("Portion of salary to save, as a decimal: "))
salary_saved = monthly_salary * portion_saved

semi_annual_raise = float(input("Semi-annual raise, as a decimal: "))
months = 0
```

```python
while current_savings < down_payment:
    current_savings += current_savings * r / 12
    current_savings += salary_saved
    months += 1
    if months % 6 == 0:
        monthly_salary += monthly_salary * semi_annual_raise
        salary_saved = monthly_salary * portion_saved

print("Number of months: ", months)
```
